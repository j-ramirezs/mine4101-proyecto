## 8. [35%] Preparación de datos:

```python
#Importar librerías

import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import sklearn
import matplotlib as mplt
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
df = pd.read_csv('./Flight_Report_17.csv', encoding = "UTF-8", sep = ',')
# Se elimina las columnas 'Unnamed' que pueden surgir en ciertas ocasiones
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.shape
```

```python
df.info()
```

```python
df.sample(10)
```

### 8.1. Selección de variables

Previo al procesamiento de datos, se eliminan las variables que no serán relevantes para los análisis seún el negocio. Además, dado que el problema o la solución que se está planteando es predecir el tiempo de vuelo o consumo de batería del dron, también se eliminan las variables 'Landing', pues al momento del despegue, no se contará con dichas variables.

```python
df = df[['Pilot-in-Command', 'Air Seconds', 'Above Sea Level (Meters)', 'Drone Type', 'Takeoff Bat %', 'Takeoff mAh', 'Takeoff Volts', 'Landing Bat %', 'Max Altitude (Meters)', 'Total Mileage (Kilometers)']]
```

```python
df.head()
```

### 8.2. Análisis y corrección de valores nulos

```python
df.isna().sum()/len(df)
```

De acuerdo a lo anterior, podemos observar que hay 3 campos con valores nulos, de los cuales 1 tiene un procentaje mayor al 30%, para este caso es impensable realizar algún proceso de completado dado que es un procentaje demasiado alto, por lo que se opta por eliminar esta columna.

```python
df = df[['Pilot-in-Command', 'Air Seconds', 'Above Sea Level (Meters)', 'Drone Type', 'Takeoff Bat %', 'Takeoff Volts', 'Landing Bat %', 'Max Altitude (Meters)', 'Total Mileage (Kilometers)']]
```

Ahora, para los 2 campos restantes con valores nulos, se realizará una imputación de datos

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].hist(df[['Above Sea Level (Meters)']], bins=20, color='blue', alpha=0.7)
axs[0].set_title('Above Sea Level (Meters)')
axs[1].hist(df[['Takeoff Volts']], bins=20, color='green', alpha=0.7)
axs[1].set_title('Takeoff Volts')
plt.tight_layout()
plt.show()
```

Dado que las distribuciones no son completamente normales, se hará la imputación de los datos por la mediana.

```python
df['Above Sea Level (Meters)'].fillna(df['Above Sea Level (Meters)'].median(), inplace=True)
df['Takeoff Volts'].fillna(df['Takeoff Volts'].median(), inplace=True)
```

```python
df.isna().sum()/len(df)
```

### 8.3. Codificación de variables categóricas

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
variables_categoricas = ['Pilot-in-Command', 'Drone Type']
```

```python
df_model = df.copy()
df_model.head()
```

```python
encoder = OneHotEncoder(sparse=False, drop='first')
encoder.fit(df_model[variables_categoricas])
encoded_data = encoder.transform(df_model[variables_categoricas])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(variables_categoricas))
df_model = pd.concat([df_model, encoded_df], axis=1)
df_model.drop(variables_categoricas, axis=1, inplace=True)
```

```python
df_model.head()
```

### 8.4. Normalización de valores

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df_model)
df_model = pd.DataFrame(scaler.transform(df_model), columns=df_model.columns)
```

```python
df_model.head()
```

## 10. [10%] Construcción del modelo

Dado que tenemos dos casos, para el caso 1, los valores de Y o las predicciones serán los valores del campo correspondiente al tiempo de vuelo ("Air seconds") el data set no contará con la variable "Landing Bat %". Para el segundo caso la variable a predecir será "Landing Bat %".

```python
df_caso1 = df_model.copy()
df_caso1.head()
```

```python
df_caso1 = df_caso1.drop('Landing Bat %', axis=1)
df_caso1.head()
```

```python
df_caso2 = df_model.copy()
df_caso2.head()
```

### 10.1 Partición de los datos

```python
features_1 = df_caso1.columns.tolist()
features_1
```

```python
features_1.remove('Air Seconds')
features_1
```

```python
features_2 = df_caso2.columns.tolist()
features_2.remove('Landing Bat %')
features_2
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(df_caso1[features_1], df_caso1["Air Seconds"], test_size=0.3, random_state=1)
```

```python
X_train_2, X_test_2, Y_train_2, Y_test_2 = train_test_split(df_caso2[features_2], df_caso2["Landing Bat %"], test_size=0.3, random_state=1)
```

```python
X_train_1.shape, Y_train_1.shape
```

```python
X_test_1.shape, Y_test_1.shape
```

```python
X_train_2.shape, Y_train_2.shape
```

```python
X_test_2.shape, Y_test_2.shape
```

### 10.2 Entrenamiento de un modelo de regresión lineal

```python
from sklearn.linear_model import LinearRegression
```

```python
regression_1 = LinearRegression()
regression_2 = LinearRegression()
```

```python
regression_1.fit(X_train_1, Y_train_1)
regression_2.fit(X_train_2, Y_train_2)
```

Interpretación de coeficientes

```python
pd.DataFrame({"columns": features_1, "coef": regression_1.coef_})
```

```python
pd.DataFrame({"columns": features_2, "coef": regression_2.coef_})
```

## 11. [20%] Evaluación del modelo:

### 11.1 Evaluación de los modelos de regresión

**Mean Absolute Error (MAE)**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("MAE Train caso 1:", np.sqrt(mean_absolute_error(Y_train_1, regression_1.predict(X_train_1))))
print("MAE Test caso 1:", np.sqrt(mean_absolute_error(Y_test_1, regression_1.predict(X_test_1))))
```

```python
print("MAE Train caso 2:", np.sqrt(mean_absolute_error(Y_train_2, regression_2.predict(X_train_2))))
print("MAE Test caso 2:", np.sqrt(mean_absolute_error(Y_test_2, regression_2.predict(X_test_2))))
```

**Root Mean Squeared Error (RMSE)**

```python
print("RMSE Train caso 1:", np.sqrt(mean_squared_error(Y_train_1, regression_1.predict(X_train_1))))
print("RMSE Test caso 1:", np.sqrt(mean_squared_error(Y_test_1, regression_1.predict(X_test_1))))
```

```python
print("RMSE Train caso 2:", np.sqrt(mean_squared_error(Y_train_2, regression_2.predict(X_train_2))))
print("RMSE Test caso 2:", np.sqrt(mean_squared_error(Y_test_2, regression_2.predict(X_test_2))))
```
